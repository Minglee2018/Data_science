```python

def clear_row_EN(s):
    a =''
    for index,items in enumerate(s):
        if items.find('District') != -1:
            items = items.replace('District','')
            items = ' Quận '+items
            s[index]=items
        if items.find('Street') != -1:
            items = items.replace('Street','')
            items = 'Đường'+items
            s[index]=items
        if items.find('City') != -1:
            items = items.replace('City','')
            items = 'Thành Phố'+items
            s[index]=items
        if items.find('Road') != -1:
            items = items.replace('Road','')
            items = 'Đường'+items
            s[index]=items
        if items.find('Province') != -1:
            items = items.replace('Province','')
            items = 'Thành phố'+items
            s[index]=items
        if items.find('Building') != -1:
            items = items.replace('Building','')
            items = 'Tòa Nhà'+items
            s[index]=items
        s[index]= s[index]
        a+=s[index]+','
    return a
```

```python
def clear_row_Keys(s):
    a =''
    for index,items in enumerate(s):
        if items.find(' Q ') != -1:
            items = items.replace('Q','Quận')
        if items.find('TP') != -1:
            items = items.replace('TP','Thành Phố')
        if items.find('QL') != -1:
            items = items.replace('QL','Quốc lộ')
        if items.find('KP') !=-1:
            items = items.replace('KP','khu phố')
        if items.find('KCN') !=-1:
            items = items.replace('KCN','khu công nghiệp ')
        if items.find( 'KV') !=-1:
            items = items.replace('KV','Khu vực ')
        if items.find('CN') !=-1:
            items = items.replace('CN','công nghiệp ')
        if items.find('SN') !=-1:
            items = items.replace('SN','số nhà ')
        if items.find('TL') !=-1:
            items = items.replace(' TL ','Thạnh Lộc ')
        if items.find('TT') !=-1:
            items = items.replace(' TT ','Thị trấn ')
        if items.find('Hà Nội')!=-1 or items.find('hà nội')!=-1 or items.find('Hà nội')!=-1 or items.find('ha noi')!=-1 or items.find('Ha Noi')!=-1:
            items=' Thành Phố Hà Nội '
        a+=items+','
    return a
```

```python
for i in range(20):
    s = labels.location[i].split(',')
    s = clear_row_EN(s)
    s = s.replace(',,,',',').replace(',,',',')
    labels.location[i] = s
    #s = s[0:len(s)-1]
    print(s)
```
