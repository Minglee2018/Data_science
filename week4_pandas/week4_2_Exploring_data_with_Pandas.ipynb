## Using iloc to select by integer position

To select data by integer position using pandas we use the Dataframe.iloc[] method and the Series.iloc[] method. It's easy to get loc[] and iloc[] confused at first, but the easiest way is to remember the first letter of each method:

- loc: lable based selection

- iloc: integer position based selection

```python
# Select the first row
f500.iloc[0]
```

```python
# Select the second row
f500.iloc[1, :]
```

```python
# Select the first column
f500.iloc[:, 0]
```

```python
# Select multiple rows using slicing
f500[0:3]
```

```python
# Select multiple rows using slicing
f500.iloc[0:3]
```

```python
# Select multiple rows, and columns using slicing
f500.iloc[0:3, 0:7]
```

## Using Loops with pandas

We create a dictionary that records the top employer in each country. First, how to get the top employer in each country.

```python
# Top employers in USA
f500.loc[f500['country'] == 'USA', 'employees'].sort_values(ascending=False).head()
```

```python
# Top employers in China
f500.loc[f500['country'] == 'China', 'employees'].sort_values(ascending=False).head()
```

We don't want to change the country manually to get the result. How to get and store the results of all country ? First, we get the list of all country

```python
country_list = f500['country'].unique()
```

Then, we loop through each country in the country_list and using the above code to get the result and store it in a dictionary.

```python
top_employer = {}
for country in country_list:
    top_employer[country] = f500.loc[f500['country'] == country, 'employees'].sort_values(ascending=False).head().index[0]
top_employer
```

### Calculating Return on Assets by Country

First, we create a column that calculate the return on assest.

```python
f500['roa'] = f500['profits'] / f500['assets']
```

```python
f500.head()
```

Then, we create a dictionary where 

- Dictionary keys is the sector name.

- Dictionary values is the company name with the highest ROA value from that sector.

```python
sector_list = f500['sector'].unique()
```

```python
highest_roa_by_sector = {}
for sector in sector_list:
    sorted_roa = f500.loc[f500['sector'] == sector, 'roa'].sort_values(ascending=False)
    highest_roa_by_sector[sector] = list(sorted_roa.items())[0]
highest_roa_by_sector
```
